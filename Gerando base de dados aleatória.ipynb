```python
# Só iremos precisar do Pandas e do Random, também poderia ser criado com Numpy mas acho mais simples assim
import pandas as pd
import random as rd
```

```python
'''
Como o n de observações será grande, vamos criar uma variável com lista vazia e depois usar o For para gerar tantos valores
aleatórios quanto necessários (300, neste caso). Depois do laço, usamos a função rename do Pandas para renomear a coluna porque
by default ela tem como nome o int(0). Não consegui pensar uma forma melhor de nomear a coluna dentro do laço, então esse foi
o caminho que pareceu mais razoável pra mim.
'''
filmes = []

for n in range(300):
    filmes.append(rd.randint(1,20))
    df = pd.DataFrame(filmes)
    
df.rename(columns={0:'N de filmes'}, inplace=True)
df
```

```python
'''
Agora, vamos criar a coluna com os tipos de filmes. Para isso, criei a variável e passei uma lista com strings de tipos de
filmes. Como a função .randint retorna apenas números inteiros, usamos a .choices para escolher uma string aleatória dentro da
lista que criamos, com k sendo o número de escolhas a serem feitas (neste caso 300, para igualar o n). Pronto!
'''
categorias = ['Drama', 'Romance', 'Comédia', 'Ação', 'Terror', 'Suspense']
df['Tipo mais assistido'] = rd.choices(categorias, k=300)

df
```

```python
'''
Agora, só exportar o arquivo para .csv também utilizando uma função do Pandas. É importante utilizar o encoding, porque o Excel
não lê os caracteres especiais (sinais gráficos e acentos) e isso pode causar problemas na leitura dos dados. Outra função a ser
utilizada é a index=False, para que o csv gerado seja somente com as colunas.
'''
df.to_csv('dados_filmes.csv', encoding='utf-8-sig', index=False)
```
